# **check which area got most of the properties**

```python
#select latitude and longitude columns from the data
Airbnb[['latitude','longitude']]
```

```python
# with the help of the scatter plot lets find the density of neighbourhood across the location
plt.figure(figsize=(20,10))
sns.set(style='whitegrid')
axis_5=sns.scatterplot(data=Airbnb, y='latitude', x='longitude', hue='neighbourhood_group')
axis_5.set_title('density of rooms')
axis_5.set_xlabel('longitude')
axis_5.set_ylabel('latitude')
plt.show()
```

# **Analysis on roomtype on the basis of areas** 

```python
#The data of room type within the neighbourhood groups
roomtype_within_neighbourhood = Airbnb.groupby('neighbourhood_group')['room_type'].value_counts()
roomtype_within_neighbourhood.head(20)
```

```python
#Lets plot the above data with the help of countplot
plt.figure(figsize=(20,10))
sns.set(style='whitegrid')
sns.countplot(data=Airbnb, x='neighbourhood_group', hue='room_type',palette='Paired')
plt.title('Differentiation of Room types in different Neighbourhood areas')
plt.show()
```

# **which area got maximum reviews?**

```python
# Area analysis on the basis of number_of_reviews
area_reviews = Airbnb.groupby(['neighbourhood_group'])['number_of_reviews'].max().reset_index()
area_reviews
```

```python
# create the bar chart from the above data
plt.figure(figsize=(10,5))
sns.barplot(data=area_reviews,x='neighbourhood_group',y='number_of_reviews',palette='Set2')
plt.title('neighbourhood_group vs number_of_reviews')
plt.xlabel('neighbourhood_group')
plt.ylabel('number_of_reviews')
plt.show()
```

# **what can we learn from price predictions on the basis of reviews?**

```python
# price analsis on the basis number_of_reviews
price_reviews = Airbnb.groupby(['price'])['number_of_reviews'].max().reset_index()
price_reviews_1=price_reviews.head(10)
price_reviews_1
```

```python
# create the lineplot from the above data
plt.figure(figsize=(15,5))
sns.lineplot(data=price_reviews_1,x='price',y='number_of_reviews')
plt.title('price vs number_of_reviews')
plt.xlabel('price')
plt.ylabel('number_of_reviews')
plt.show() 
```

# **which hosts are the most busiest hosts and why?**

```python
# Finding buisest hosts
busiest_hosts = Airbnb.groupby(['host_name','neighbourhood_group'])['host_id'].value_counts().reset_index(name='counts')
busy_hosts = busiest_hosts.sort_values(by=['counts'], ascending=False).head(10)
busy_hosts
```

```python
# create a barplot,of a given data
plt.figure(figsize=(15,5))
sns.barplot(data=busy_hosts,x='host_name',y='counts')
plt.xticks(rotation=45)
plt.title('Busiest Hosts')
plt.xlabel('host_name')
plt.ylabel('counts')
plt.show() 
```

# **Is there any noticeable difference of traffic among different areas and what could be the reason for it?**

```python
traffic_areas = Airbnb.groupby(['neighbourhood_group','room_type'])['minimum_nights'].count().reset_index()
traffic_areas_1 = traffic_areas.sort_values(by='minimum_nights', ascending=False)
traffic_areas_1
```

```python
#create boxplot from the given data
plt.figure(figsize=(15,7))
sns.boxplot(data=traffic_areas_1, x='room_type', y='minimum_nights', hue='room_type')
plt.title('Traffic among different areas')
plt.xlabel('room_type')
plt.ylabel('minimum_number_of_nights stayed')
plt.show()
```
